### Análisis Exploratorio del dataset 

### Plan de preprocesamiento por cada atributo
- def-instance -> drop (optional)
- color -> drop
- state -> apply one-hot encoding (this generates 30 new attributes)
- control -> has 1 nan: fill with mode and apply one-hot encoding (this generates 5 new attributes)
- no-of-students -> map to ordinal integers (5 unique values)
- male:female -> map to single integer (e.g. male ratio)
- student:faculty -> has 1 nan: fill with median and map to single integer (e.g. student ratio)
- sat-verbal -> has 10 nan: group by percent-admittance ranges and fill with median of each group
- sat-math -> has 10 nan: group by percent-admittance ranges and fill with median of each group
- expenses -> map to ordinal integers (4 unique values)
- percent-financial-aid -> has 2 nan: fill with median
- no-applicants -> map to ordinal integers (6 unique values)
- percent-admittance -> OK
- percent-enrolled -> OK
- academics-scale:1-5 -> OK
- social-scale:1-5 -> OK
- quality-of-life:1-5 -> OK
- academic-emphasis -> split by ':', count unique values, create columns for each value and fill with 0 or 1
- location -> has 2 nan: fill with mode and apply one-hot encoding (generates 4 new attributes)

```python
# get copy of dataframe
df = univ_data
```

```python
# delete nan rows
df = df[pd.notnull(df['def-instance'])]
# delete nan columns
df = df.drop('colors', 1)
# def-instance -> drop (optional)
df = df.drop('def-instance', 1)
```

```python
# state -> apply one-hot encoding (this generates 30 new attributes)
df = df.join(pd.get_dummies(df['state']))
df = df.drop('state', 1)
```

```python
# control -> has 1 nan: fill with mode and apply one-hot encoding (this generates 5 new attributes)
df['control'] = df['control'].fillna(df['control'].mode()[0])
df = df.join(pd.get_dummies(df['control']))
df = df.drop('control', 1)
```

```python
# no-of-students -> map to ordinal integers (5 unique values)
print(df['no-of-students'].unique())
replace_dict = {
    'thous:5-': 0,
    'thous:5-10': 1,
    'thous:10-15': 2,
    'thous:15-20': 3,
    'thous:20+': 4
}
df['no-of-students'] = df['no-of-students'].replace(replace_dict)
```

```python
# male:female -> map to single integer (e.g. male ratio)
df['male:female'] = df['male:female'].apply(lambda x: int(x.split(':')[1]))
```

```python
# student:faculty -> has 1 nan: fill with median and map to single integer (e.g. student ratio)
df['student:faculty'] = df['student:faculty'].apply(
    lambda x: int(x.split(':')[1]) if not pd.isnull(x) else pd.np.nan)
df['student:faculty'] = df['student:faculty'].fillna(df['student:faculty'].median())
```

```python
# sat-verbal -> has 10 nan: group by percent-admittance ranges and fill with median of each group
df['range-percent-admittance'] = pd.cut(df['percent-admittance'].astype(int), 
                                        bins=[-1,20,40,60,80,100], labels=False)
labels = pd.np.array(['0-20','21-40','41-60','61-80','81-100'])
df['range-percent-admittance'] = labels[df['range-percent-admittance']]
df['sat-verbal'] = df.groupby('range-percent-admittance')['sat-verbal'].transform(
    lambda x: x.fillna(x.dropna().astype(int).median())).astype(int)
df = df.drop('range-percent-admittance', 1)
```

```python
# sat-math -> has 10 nan: group by percent-admittance ranges and fill with median of each group
df['range-percent-admittance'] = pd.cut(df['percent-admittance'].astype(int), 
                                        bins=[-1,20,40,60,80,100], labels=False)
labels = pd.np.array(['0-20','21-40','41-60','61-80','81-100'])
df['range-percent-admittance'] = labels[df['range-percent-admittance']]
df['sat-math'] = df.groupby('range-percent-admittance')['sat-math'].transform(
    lambda x: x.fillna(x.dropna().astype(int).median())).astype(int)
df = df.drop('range-percent-admittance', 1)
```

```python
# expenses -> map to ordinal integers (4 unique values)
print(df['expenses'].unique())
replace_dict = {
    'thous$:4-': 0,
    'thous$:4-7': 1,
    'thous$:7-10': 2,
    'thous$:10+': 3,
}
df['expenses'] = df['expenses'].replace(replace_dict)
```

```python
# percent-financial-aid -> has 2 nan: fill with median
df['percent-financial-aid'] = df['percent-financial-aid'].fillna(
    df['percent-financial-aid'].dropna().astype(int).median()).astype(int)
```

```python
# no-applicants -> map to ordinal integers (6 unique values)
print(df['no-applicants'].unique())
replace_dict = {
    'thous:4-': 0,
    'thous:4-7': 1,
    'thous:7-10': 2,
    'thous:10-13': 3,
    'thous:13-17': 4,
    'thous:17+': 5
}
df['no-applicants'] = df['no-applicants'].replace(replace_dict)
```

```python
# percent-admittance -> OK
# percent-enrolled -> OK
# academics-scale:1-5 -> OK
# social-scale:1-5 -> OK
# quality-of-life-scale:1-5 -> OK
df['percent-admittance'] = df['percent-admittance'].astype(int)
df['percent-enrolled'] = df['percent-enrolled'].astype(int)
df['academics-scale:1-5'] = df['academics-scale:1-5'].astype(int)
df['social-scale:1-5'] = df['social-scale:1-5'].astype(int)
df['quality-of-life-scale:1-5'] = df['quality-of-life-scale:1-5'].astype(int)
```

```python
# academic-emphasis -> split by ':', count unique values, create columns for each value and fill with 0 or 1
academics = filter(None, set(sum([x.split(':') for x in df['academic-emphasis'].unique()], [])))
df = pd.concat([df, pd.DataFrame(columns=academics)])
df[academics] = df[academics].fillna(0)
for academic in academics:
    df[academic] = df['academic-emphasis'].apply(lambda x: int(len(x)>0 and academic in x.split(':')))
df = df.drop('academic-emphasis', 1)
```

```python
# location -> has 2 nan: fill with mode and apply one-hot encoding (generates 4 new attributes)
df['location'] = df['location'].fillna(df['location'].mode()[0])
df = df.join(pd.get_dummies(df['location']))
df = df.drop('location', 1)
```
